```python
import matplotlib.pyplot as plt
```

```python
def Midpoint(x1,y1,x2,y2):
    dx = x2-x1
    dy = y2-y1
    
    p = 2*dy - dx
    
    x,y = x1,y1
    
    line_x = [x]
    line_y = [y]
    
    while(x < x2):
        if p < 0:
            x += 1
            line_x.append(x)
            line_y.append(y)
            p = p + 2*dy
            
        else:
            x += 1
            y += 1
            line_x.append(x)
            line_y.append(y)
            p = p + 2*dy - 2*dx
            
    return line_x, line_y


x1, y1 = 5,5
x2,y2 = 15,15

line_x, line_y = Midpoint(x1,y1,x2,y2)


plt.plot(line_x, line_y, marker='o')
plt.title("Mid Point Line Drawing Algorithm")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
