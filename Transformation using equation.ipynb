```python

import matplotlib.pyplot as plt
import numpy as np

def plot_polygon(polygon, title):
    # Plot the polygon
    x, y = zip(*polygon + [polygon[0]])  # Closing the polygon
    plt.plot(x, y, label='Original')

    # Add labels and title
    plt.title(title)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')

    # Set equal scaling for better visualization
    plt.axis('equal')

    # Display the legend
    plt.legend()

    # Show the plot
    plt.show()
```

```python
def translate_polygon(polygon, tx, ty):
    # Translate each point of the polygon by (tx, ty)
    translated_polygon = []
    for x, y in polygon:
        translated_polygon.append((x + tx, y + ty))
    return translated_polygon
```

```python
# Define the polygon (example: square)
polygon = [(0, 0), (1, 0), (1, 1), (0, 1)]

# Plot the original polygon
plot_polygon(polygon, 'Original Polygon')

# Translation
translation = (1, 2)
translated_polygon = translate_polygon(polygon, *translation)
plot_polygon(translated_polygon, 'Translation')
```

```python
def reflect_polygon(polygon, axis):
    # Reflect the polygon across the specified axis ('x' or 'y')
    reflected_polygon = []
    for x, y in polygon:
        if axis == 'x':
            reflected_polygon.append((x, -y))
        else:
            reflected_polygon.append((-x, y))
    return reflected_polygon
```

```python
def shear_polygon(polygon, shx, shy):
    # Shear each point of the polygon by (shx, shy)
    sheared_polygon = []
    for x, y in polygon:
        sheared_polygon.append((x + shx * y, y + shy * x))
    return sheared_polygon
```

```python
# Define the polygon (example: square)
polygon = [(0, 0), (1, 0), (1, 1), (0, 1)]

# Plot the original polygon
plot_polygon(polygon, 'Original Polygon')
```

```python
# Shearing
shearing_factors = (0.5, 0)
sheared_polygon = shear_polygon(polygon, *shearing_factors)
plot_polygon(sheared_polygon, 'Shearing')
```

```python

import matplotlib.pyplot as plt
import numpy as np

def plot_polygon(polygon, title):
    # Plot the polygon
    x, y = zip(*polygon + [polygon[0]])  # Closing the polygon
    plt.plot(x, y, label='Original')

    # Add labels and title
    plt.title(title)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')

    # Set equal scaling for better visualization
    plt.axis('equal')

    # Display the legend
    plt.legend()

    # Show the plot
    plt.show()
```

```python
def rotate_polygon(polygon, angle):
    # Rotate each point of the polygon by the specified angle (in degrees) around the origin
    angle_rad = np.radians(angle)
    rotated_polygon = []
    for x, y in polygon:
        new_x = x * np.cos(angle_rad) - y * np.sin(angle_rad)
        new_y = x * np.sin(angle_rad) + y * np.cos(angle_rad)
        rotated_polygon.append((new_x, new_y))
    return rotated_polygon
```

```python
# Define the polygon (example: square)
polygon = [(0, 0), (1, 0), (1, 1), (0, 1)]

# Plot the original polygon
plot_polygon(polygon, 'Original Polygon')
```

```python
# Rotation
rotation_angle = 45  # in degrees
rotated_polygon = rotate_polygon(polygon, rotation_angle)
plot_polygon(rotated_polygon, 'Rotation')
```
